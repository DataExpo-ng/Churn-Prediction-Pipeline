```python
# Step 1 - iniitalize the s3 client
s3_client = boto3.client('s3')

# step 2 - configuration 
bucket_name = "mo-churn-bucket"
print("connection successfully")


# step 3 - upload the csv file to the s3 bucket
def upload_file_to_s3(file_path=None, bucket_name=None, s3_key=None):
    
    # step 1 - defining the file path, bucket_name, and s3_key
    file_path = "C:/Users/Muham/OneDrive/Documents/GitHub/Churn-Prediction-Pipeline/Telco-customer-churn.csv"
    bucket_name = "mo-churn-bucket"
    s3_key = "raw/Telco-customers-churn.csv"

    # step 2 - uploading the file
    s3_client.upload_file(file_path, bucket_name, s3_key)
    print(f"File uploaded to s3://{bucket_name}/{s3_key} sucessfully")


# step 4 - call the function
upload_file_to_s3()
```
